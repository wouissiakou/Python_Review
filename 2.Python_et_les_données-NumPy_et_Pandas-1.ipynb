### <left> <b> <span style="color:orange;"> 0.4. Array Slicing : Accéder aux sous-tableaux </span> </b></left>


Tout comme nous pouvons utiliser des crochets pour accéder aux éléments individuels du tableau, nous pouvons également les utiliser pour accéder aux sous-tableaux avec la notation *tranche*, marquée par le caractère deux-points (``:``).
La syntaxe de découpage NumPy suit celle de la liste Python standard ; pour accéder à une tranche d'un tableau ``x``, utilisez ceci :

```python
x[début:arrêt:étape]  # x[start:stop:step]
```

Si l'un d'entre eux n'est pas spécifié, il prend par défaut les valeurs ``start=0``, ``stop=``*``size of dimension``*, ``step=1``.

Nous verrons comment accéder aux sous-tableaux dans une dimension et dans plusieurs dimensions.

#### Sous-tableaux unidimensionnels

```python
x = np.range(10)
X

x[:5] # cinq premiers éléments

x[5:] # éléments après l'index 5

x[4:7] # sous-tableau du milieu

x[::2] # tous les autres éléments

x[1::2] # tous les autres éléments, à partir de l'index 1
```

Un cas potentiellement déroutant est lorsque la valeur ``step`` est négative. Dans ce cas, les valeurs par défaut pour ``start`` et ``stop`` sont permutées. Cela devient un moyen pratique d'inverser un tableau :

```python
x[::-1] # tous les éléments, inversés

x[5::-2] # inversé tous les autres à partir de l'index 5
```

#### Sous-tableaux multidimensionnels

Les tranches multidimensionnelles fonctionnent de la même manière, avec plusieurs tranches séparées par des virgules.
Par example:

```python
x2

x2[:2, :3] # deux lignes, trois colonnes

x2[:3, ::2] # toutes les lignes, toutes les autres colonnes
```
Enfin, les dimensions des sous-tableaux peuvent même être inversées :"""

```python
x2[::-1, ::-1]
```


### <left> <b> <span style="color:orange;"> 0.5. Accéder aux lignes et aux colonnes du tableau </span> </b></left>


Une routine couramment nécessaire consiste à accéder à des lignes ou des colonnes uniques d'un tableau.
Cela peut être fait en combinant l'indexation et le découpage, en utilisant une tranche vide marquée par un seul deux-points (``:``):


```python
print(x2[:, 0]) # première colonne de x2

print(x2[0, :]) # première ligne de x2
```

Dans le cas d'un accès aux lignes, la tranche vide peut être omise pour une syntaxe plus compacte :"""

```python
print(x2[0]) # équivalent à x2[0, :]
```




### <left> <b> <span style="color:orange;"> 0.6. Remodelage des tableaux </span> </b></left>


Un autre type d'opération utile est le remodelage des tableaux. La manière la plus flexible de le faire est d'utiliser la méthode ``reshape``. Par exemple, si vous voulez mettre les nombres 1 à 9 dans une grille $3\times 3$, vous pouvez faire ce qui suit :


```python
grille = np.arange(1, 10).reshape((3, 3))
print(grille)
```

Notez que pour que cela fonctionne, la taille du tableau initial doit correspondre à la taille du tableau remodelé.
Dans la mesure du possible, la méthode ``reshape`` utilisera une vue sans copie du tableau initial, mais avec des mémoires tampons non contiguës, ce n'est pas toujours le cas.

Un autre modèle de remodelage courant est la conversion d'un tableau unidimensionnel en une matrice de lignes ou de colonnes bidimensionnelle.
Cela peut être fait avec la méthode ``reshape``, ou plus facilement en utilisant le mot-clé ``newaxis`` dans une opération slice :


```python
x = np.arrat([1, 2, 3])

# vecteur de ligne via remodelage
x.reshape((1, 3))

# vecteur ligne via newaxis
x[np.newaxis, :]

# vecteur de colonne via remodelage
x.reshape((3, 1))

# vecteur de colonne via newaxis
x[:, np.newaxis]
```

Nous verrons souvent ce type de transformation dans les lecons a venir.

### <left> <b> <span style="color:orange;"> 0.7. Concaténation et fractionnement de tableaux </span> </b></left>



Toutes les routines précédentes fonctionnaient sur des tableaux uniques. Il est également possible de combiner plusieurs tableaux en un seul et, inversement, de diviser un seul tableau en plusieurs tableaux. Nous reviendrons sur ces opérations ici.

#### <left> <b> <span style="color:green;"> 0.7.0. Concaténation de tableaux </span> </b></left>

La concaténation, ou la jonction de deux tableaux dans NumPy, est principalement réalisée à l'aide des routines ``np.concatenate``, ``np.vstack`` et ``np.hstack``.
``np.concatenate`` prend un tuple ou une liste de tableaux comme premier argument, comme nous pouvons le voir ici :
"""


```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np. concatenate([x, y])
```

Vous pouvez également concaténer plus de deux tableaux à la fois :


```python
z = [99, 99, 99]
print(np. concatenate([x, y, z]))
```


Il peut également être utilisé pour les tableaux à deux dimensions :

```python
grille = np.array([[1, 2, 3], [4, 5, 6]])

# concaténer le long du premier axe
np.concatenate([grille, grille])

# concaténer le long du deuxième axe (indexé à zéro)
np. concatenate([grille, grille], axe=1)
```


Pour travailler avec des tableaux de dimensions mixtes, il peut être plus clair d'utiliser les fonctions ``np.vstack`` (pile verticale) et ``np.hstack`` (pile horizontale):


```python
x = np.array([1, 2, 3])
grille = np.array([[9, 8, 7], [6, 5, 4]])

# empiler verticalement les tableaux
np.vstack([x, grille])

# empiler horizontalement les tableaux
y = np.array([[99],
              [99]])
np.hstack([grille, y])
```


De même, ``np.dstack`` empilera les tableaux le long du troisième axe.




#### <left> <b> <span style="color:green;"> 0.7.1. Fractionnement des tableaux </span> </b></left>


Le contraire de la concaténation est le fractionnement, qui est implémenté par les fonctions ``np.split``, ``np.hsplit`` et ``np.vsplit``. Pour chacun d'entre eux, on peut passer une liste d'indices donnant les points de partage :

```python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

Remarquez que *N* points de partage, conduit à *N + 1* sous-tableaux. Les fonctions associées ``np.hsplit`` et ``np.vsplit`` sont similaires :


```python
grille = np.arange(16).reshape((4, 4))
la grille

superieur, inferieur = np.vsplit(grille, [2])
print(superieure)
print(inferieure)

gauche, droite = np.hsplit(grille, [2])
print(gauche)
print(droite)
```

De même, ``np.dsplit`` divisera les tableaux le long du troisième axe.

### <left> <b> <span style="color:orange;"> 0.8. Calcul sur les tableaux NumPy : fonctions universelles et autres </span> </b></left>



Le calcul sur les tableaux NumPy peut être très rapide ou très lent. La clé pour le rendre rapide est d'utiliser des opérations *vectorisées*, généralement implémentées via les *fonctions universelles* de NumPy (ufuncs).

Cette section motive le besoin des ufuncs de NumPy, qui peuvent être utilisés pour rendre les calculs répétés sur les éléments du tableau beaucoup plus efficaces.  Il présente ensuite bon nombre des ufuncs arithmétiques les plus courantes et les plus utiles disponibles dans le package NumPy.


#### <left> <b> <span style="color:green;"> 0.8.0. La lenteur des boucles </span> </b></left>

Par exemple, imaginons que nous ayons un tableau de valeurs et que nous aimerions calculer l'inverse de chacune.

#### <left> <b> <span style="color:green;"> 0.8.1. Présentation des UFuncs </span> </b></left>

Pour de nombreux types d'opérations, NumPy fournit une interface pratique dans ce type de routine compilée et typée statiquement. C'est ce qu'on appelle une opération *vectorisée*. Cela peut être accompli en effectuant simplement une opération sur le tableau, qui sera ensuite appliquée à chaque élément.



##### Arithmétique des tableaux
 Les ufuncs de NumPy semblent très naturels à utiliser car ils utilisent les opérateurs arithmétiques natifs de Python.
L'addition, la soustraction, la multiplication et la division standard peuvent toutes être utilisées :


```python
x = np.range(4)
print("x=",x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x // 2 =", x // 2) # division entiere
```

Il existe également un ufunc unaire pour la négation, un opérateur ``**`` pour l'exponentiation et un opérateur ``%`` pour le module :

```python
print("-x     = ", -x)
print("x ** 2 = ", x ** 2)
print("x % 2  = ", x % 2)
```


De plus, ceux-ci peuvent être enchaînés comme bon vous semble, et l'ordre standard des opérations est respecté :
```python
-(0.5*x + 1) ** 2
```

Chacune de ces opérations arithmétiques sont simplement des enveloppes pratiques autour de fonctions spécifiques intégrées à NumPy ; par exemple, l'opérateur ``+`` est un wrapper pour la fonction ``add`` :

```python
np.add(x, 2)
```

##### Exposants et logarithmes
Un autre type d'opération courant disponible dans une ufunc NumPy sont les exponentielles :

```python
x = [1, 2, 3]
print("x     =", x)
print("e^x   =", np.exp(x))
print("2^x   =", np.exp2(x))

```

L'inverse des exponentielles, les logarithmes, sont également disponibles.
Le ``np.log`` de base donne le logarithme népérien ; si vous préférez calculer le logarithme en base 2 ou le logarithme en base 10, ceux-ci sont également disponibles :

```python
x = [1, 2, 4, 10]

print("ln(x)    =", np.log(x))
print("log2(x)  =", np.log2(x))
```

Il existe également des versions spécialisées qui sont utiles pour maintenir la précision avec une très petite entrée :



```python

```

Lorsque ``x`` est très petit, ces fonctions donnent des valeurs plus précises que si le ``np.log`` brut ou ``np.exp`` devait être utilisé.

#### <left> <b> <span style="color:green;"> 0.8.3. Diffusion (-- Broadcasting) </span> </b></left>





Nous avons vu dans la section précédente comment les fonctions universelles (ou ufuncs) de NumPy peuvent être utilisées pour *vectoriser* les opérations et ainsi supprimer les boucles Python lentes. Un autre moyen de vectoriser les opérations consiste à utiliser la fonctionnalité *broadcasting* (ou diffusion) de NumPy.

La diffusion est simplement un ensemble de règles pour appliquer des ufuncs binaires (par exemple, addition, soustraction, multiplication, etc.) sur des tableaux de différentes tailles.

#### Présentation de la diffusion

Rappelez-vous que pour des tableaux de même taille, les opérations binaires sont effectuées élément par élément :

```python
importer numpy as np

a = np.array([0, 1, 2])
b = np.array([5, 5, 5])
a + b
```

La diffusion permet d'effectuer ces types d'opérations binaires sur des tableaux de tailles différentes. Par exemple, nous pouvons tout aussi facilement ajouter un scalaire (considérez-le comme un tableau à zéro dimension) à un tableau :


```python
a + 5
```


Nous pouvons considérer cela comme une opération qui étend ou duplique la valeur ``5`` dans le tableau ``[5, 5, 5]``, et ajoute les résultats. L'avantage de la diffusion de NumPy est que cette duplication de valeurs n'a pas réellement lieu, mais c'est un modèle mental utile lorsque nous pensons à la diffusion. Nous pouvons également étendre cela aux tableaux de dimension supérieure. Observez le résultat lorsque nous ajoutons un tableau unidimensionnel à un tableau bidimensionnel :

```python
M = np.ones((3, 3))
M
M + a
```

Ici, le tableau unidimensionnel ``a`` est étiré ou diffusé sur la deuxième dimension afin de correspondre à la forme de ``M``.
Bien que ces exemples soient relativement faciles à comprendre, des cas plus compliqués peuvent impliquer la diffusion des deux tableaux. Considérez l'exemple suivant :

```python
a = np.range(3)
b = np.arange(3)[:, np.newaxis]

print(a)
print(b)

a + b
```

Tout comme avant nous avons étiré ou diffusé une valeur pour qu'elle corresponde à la forme de l'autre, ici nous avons étiré *les deux* ``a`` et ``b`` pour correspondre à une forme commune, et le résultat est un deux- tableau dimensionnel. La géométrie de ces exemples est visualisée dans la [figure suivante](https://numpy.org/doc/stable/user/basics.broadcasting.html).



#### Règles de diffusion

La diffusion dans NumPy suit un ensemble de règles strictes pour déterminer l'interaction entre les deux tableaux :

##### - Règle 1 : Si les deux tableaux diffèrent par leur nombre de dimensions, la forme de celui qui a le moins de dimensions est *rembourrée* avec des uns sur son côté avant (gauche).
##### - Règle 2 : Si la forme des deux tableaux ne correspond à aucune dimension, le tableau dont la forme est égale à 1 dans cette dimension est étiré pour correspondre à l'autre forme.
##### - Règle 3 : Si dans n'importe quelle dimension les tailles ne sont pas d'accord et qu'aucune n'est égale à 1, une erreur est générée.

Pour rendre ces règles claires, considérons quelques exemples en détail.

###### Exemple de diffusion 1
Regardons l'ajout d'un tableau à deux dimensions à un tableau à une dimension :


```python
M = np.ones((2, 3))
a = np.range(3)
```


Considérons une opération sur ces deux tableaux. La forme des tableaux est

 ``M.shape = (2, 3)``           ``a.shape = (3,)``

Nous voyons par la règle 1 que le tableau ``a`` a moins de dimensions, nous le remplissons donc à gauche avec des uns :

``M.shape -> (2, 3)`` ``a.shape -> (1, 3)``


Par la règle 2, nous voyons maintenant que la première dimension est en désaccord, nous étirons donc cette dimension pour qu'elle corresponde :
``M.forme -> (2, 3)``
``une.forme -> (2, 3)``
 Les formes correspondent, et nous voyons que la forme finale sera ``(2, 3)`` :

```python
M + a
```


###### Exemple de diffusion 2

Examinons un exemple où les deux tableaux doivent être diffusés :

```python
a = np.arange(3).reshape((3, 1))
b = np.range(3)
```

 Encore une fois, nous allons commencer par écrire la forme des tableaux :
``a.shape = (3, 1)``             ``b.shape = (3,)``

La règle 1 dit que nous devons remplir la forme de ``b`` avec des uns :
``a.shape -> (3, 1)``                       ``b.shape -> (1, 3)``



Et la règle 2 nous dit que nous mettons à niveau chacun de ceux-ci pour qu'ils correspondent à la taille correspondante de l'autre tableau : ``a.shape -> (3, 3)``                    ``b.shape -> (3, 3)``

Parce que le résultat correspond, ces formes sont compatibles. On peut voir ça ici :

```python
a + b
```



#### La diffusion en pratique

Les opérations de diffusion sont au cœur de nombreux exemples.
Nous allons maintenant jeter un œil à quelques exemples simples où ils peuvent être utiles.

##### Centrage d'un tableau

```python
Xmoyen

#Pour revérifier que nous avons fait cela correctement, nous pouvons vérifier que le tableau centré a une moyenne proche de zéro :

```

#### <left> <b> <span style="color:green;"> 0.8.4. Comparaisons, masques et logique booléenne </span> </b></left>

Cette section couvre l'utilisation des masques booléens pour examiner et manipuler les valeurs dans les tableaux NumPy.
Le masquage apparaît lorsque vous souhaitez extraire, modifier, compter ou manipuler des valeurs dans un tableau en fonction de certains critères : par exemple, vous pouvez souhaiter compter toutes les valeurs supérieures à une certaine valeur, ou peut-être supprimer toutes les valeurs aberrantes qui sont au-dessus un certain seuil. Dans NumPy, le masquage booléen est souvent le moyen le plus efficace d'accomplir ces types de tâches.


##### Opérateurs de comparaison comme ufuncs


Nous avons introduit les ufuncs, et nous nous sommes concentrés en particulier sur les opérateurs arithmétiques. Nous avons vu que l'utilisation de ``+``, ``-``, ``*``, ``/`` et d'autres sur les tableaux conduit à des opérations élément par élément.
NumPy implémente également des opérateurs de comparaison tels que ``<`` (inférieur à) et ``>`` (supérieur à) en tant que ufuncs élément par élément. Le résultat de ces opérateurs de comparaison est toujours un tableau avec un type de données booléen.
Les six opérations de comparaison standard sont disponibles :

```python

x = np.array([1, 2, 3, 4, 5])

x < 3 # inférieur à

x > 3 # supérieur à

x <= 3 # inférieur ou égal

x >= 3 # supérieur ou égal



x != 3 # pas égal


x == 3 # égal
```

Il est également possible de faire une comparaison élément par élément de deux tableaux, et d'inclure des expressions composées :

```python
(2 * x) == (x ** 2)
```



Comme dans le cas des opérateurs arithmétiques, les opérateurs de comparaison sont implémentés comme ufuncs dans NumPy ; par exemple, lorsque vous écrivez ``x < 3``, NumPy utilise en interne ``np.less(x, 3)``.

Un résumé des opérateurs de comparaison et de leur équivalent ufunc est présenté ici :
 | Opérateur | ufunc équivalent || Opérateur | ufunc équivalent |
 |---------------|---------------------||--------- ------|---------------------|
 |``==`` |``np.equal`` ||``!=`` |``np.not_equal`` |
 |``<`` |``np.less`` ||``<=`` |``np.less_equal`` |
 |``>`` |``np.greater`` ||``>=`` |``np.greater_equal`` |

Tout comme dans le cas des ufuncs arithmétiques, celles-ci fonctionneront sur des tableaux de n'importe quelle taille et forme.
Voici un exemple en deux dimensions :


```python
rng = np.random.RandomState(0)
x = rng.randint(10, size=(3, 4))
x
x < 6
```
